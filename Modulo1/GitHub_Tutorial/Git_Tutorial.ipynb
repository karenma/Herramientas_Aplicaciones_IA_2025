## 1. Instalar y Configurar Git

### Instalando Git

- <span style="color:#22863a">**Linux:**</span> Usa tu gestor de paquetes, por ejemplo, `sudo apt-get install git` (Debian/Ubuntu) o `sudo yum install git` (Fedora/RedHat).

### Configurando Git
<div style="background-color:#eaf5ff; border-radius:8px; padding:10px; border:1px solid #b3d3ea;">

````python
# Configura tu nombre de usuario y correo de Git (ejecuta estos comandos en tu terminal, no en Python)
!git config --global user.name "Your Name"
!git config --global user.email "your.email@example.com"

# Para verificar tu configuración:
!git config --list
````

## 2. Inicializar un Repositorio Git

Para comenzar a rastrear un proyecto con Git, inicializa un repositorio en tu directorio de proyecto. Esto crea una carpeta `.git` que almacena toda la información de control de versiones.

**Comando:**

````python
#Inicializa un repositorio de git vacio
!git init
````

## 3. Verificar el Estado del Repositorio

Usa <span style="color:#005cc5;">`git status`</span> para ver qué archivos están preparados (staged), no preparados (unstaged) o no rastreados (untracked). Esto te ayuda a entender qué cambios están listos para ser confirmados y cuáles necesitan atención.

**Comando:**

````python
# Verifica el estado de tu repositorio
!git status

# Esto muestra archivos preparados, no preparados y no rastreados.
````

## 4. Agregar y Confirmar Cambios



### Preparar Archivos
Usa <span style="color:#005cc5;">`git add`</span> para preparar archivos para la confirmación. Puedes agregar archivos individuales o todos los cambios.

### Confirmar Cambios
Usa <span style="color:#005cc5;">`git commit`</span> para guardar tus cambios preparados en el repositorio. Usa <span style="color:#22863a;">`-m`</span> para agregar un mensaje de confirmación y <span style="color:#22863a;">`-a`</span> para confirmar todos los archivos rastreados (sin pasar por la etapa de preparación para archivos modificados).

**Comandos y Ejemplos:**

````python
# Prepara un archivo específico
!git add filename.txt

# Prepara todos los archivos
!git add .

# Confirma los cambios preparados con un mensaje
!git commit -m "Agregar nueva funcionalidad"

# Confirma todos los archivos rastreados (omite preparación para archivos modificados)
!git commit -am "Confirmación rápida de todos los cambios"
````

## 5. Ver el Historial de Confirmaciones


Usa <span style="color:#005cc5;">`git log`</span> para ver el historial de confirmaciones. Hay varias opciones para personalizar la salida:
- <span style="color:#22863a;">`--oneline`</span>: Muestra cada confirmación en una sola línea.
- <span style="color:#22863a;">`--graph`</span>: Muestra un gráfico de texto de la estructura de ramas.
- <span style="color:#22863a;">`--stat`</span>: Muestra estadísticas sobre lo que cambió.

**Ejemplos:**

````python
# Ver historial completo de confirmaciones
!git log

# Ver historial de confirmaciones en una línea
!git log --oneline

# Ver historial con gráfico
!git log --oneline --graph

# Ver historial con estadísticas de cambios en archivos
!git log --stat
````

## 6. Ramas y Fusiones

Las ramas te permiten trabajar en diferentes características o correcciones de forma independiente. Fusionar (merge) trae los cambios de una rama a otra.

### Crear y Listar Ramas
- <span style="color:#005cc5;">`git branch`</span>: Lista todas las ramas
- <span style="color:#005cc5;">`git branch nombre_rama`</span>: Crea una nueva rama

### Cambiar de Rama
- <span style="color:#005cc5;">`git checkout nombre_rama`</span>: Cambia a una rama (sintaxis antigua)
- <span style="color:#005cc5;">`git switch nombre_rama`</span>: Cambia a una rama (recomendado)

### Fusionar Ramas
- <span style="color:#005cc5;">`git merge nombre_rama`</span>: Fusiona la rama especificada en la rama actual

**Ejemplos:**

````python
# Lista todas las ramas
!git branch

# Crea una nueva rama
!git branch feature-branch

# Cambia a una rama (recomendado)
!git switch feature-branch

# Fusiona una rama en la rama actual
!git merge feature-branch
````

1. Create a new branch and switch to it immediately:
This is the most common and efficient way to create a new branch for developing a new feature or fixing a bug.
Código

```python
git checkout -b <new-branch-name>
```

This command performs two actions:
- It creates a new branch named <new-branch-name> based on your current branch.
- It immediately switches your working directory to the newly created branch.

2. Create a new branch without switching to it:
If you want to create a new branch but remain on your current branch, use the git branch command.
Código
```python
git branch <new-branch-name>
```

This command creates a new branch named <new-branch-name> that points to the same commit as your current branch, but your active branch remains unchanged. To switch to it later, you would then use:
Código

```python
git checkout <new-branch-name>
```

## 7. Deshacer Cambios

A veces necesitas deshacer cambios. Git ofrece varios comandos para esto:
- `git restore archivo.txt`: Restaura un archivo al último estado confirmado
- `git reset HEAD archivo.txt`: Quita un archivo del área de preparación
- `git checkout hash_commit -- archivo.txt`: Restaura un archivo de un commit específico
- `git revert hash_commit`: Crea un nuevo commit que revierte uno anterior

**Ejemplos:**

````python
# Restore a file to the last commit
!git restore filename.txt

# Unstage a file
!git reset HEAD filename.txt

# Restore a file from a specific commit
!git checkout <commit_hash> -- filename.txt

# Revert a commit (creates a new commit that undoes the changes)
!git revert <commit_hash>
````

## 8. Trabajar con Repositorios Remotos

Los remotos son versiones de tu proyecto alojadas en internet o en una red. Puedes agregar, listar y eliminar remotos.

- `git remote add origin <url>`: Agrega un repositorio remoto
- `git remote -v`: Lista los remotos
- `git remote remove origin`: Elimina un remoto

**Ejemplos:**

````python
# Add a remote repository
!git remote add origin https://github.com/username/repo.git

# List all remotes
!git remote -v

# Remove a remote
!git remote remove origin
````

## 9. Clonar Repositorios

Para copiar un repositorio remoto a tu máquina local, usa `git clone`.

**Ejemplo:**

````python
# Clone a remote repository to your local machine
!git clone https://github.com/username/repo.git
````

## 10. Subir y Descargar Cambios

- `git push`: Sube tus confirmaciones locales a un repositorio remoto
- `git pull`: Descarga y fusiona cambios de un repositorio remoto
- `git fetch`: Descarga cambios de un repositorio remoto (no fusiona)

**Ejemplos:**

````python
# Push local commits to the remote repositorygit 
!git push origin main

# Pull changes from the remote repository and merge
!git pull origin main

# Fetch changes from the remote repository (does not merge)
!git fetch origin
````

## 11. Resolver Conflictos de Fusión

Al fusionar ramas, pueden ocurrir conflictos si las mismas líneas se modifican en ambas ramas. Git marcará los archivos en conflicto y deberás resolverlos manualmente.

- Edita los archivos en conflicto para resolver las diferencias.
- Usa `git status` para ver qué archivos están en conflicto.
- Después de resolver, prepara los archivos y confirma la fusión.

**Ejemplo de Flujo:**

````python
# After a merge conflict, check status
!git status

# Edit the conflicted files to resolve conflicts
# Then stage the resolved files
!git add conflicted_file.txt

# Commit the merge after resolving conflicts
!git commit
````

## 12. Guardar Cambios Temporales (Stash)

El stash te permite guardar cambios que aún no estás listo para confirmar. Puedes volver a aplicarlos o descartarlos después.

- `git stash`: Guarda tus cambios
- `git stash list`: Lista todos los stashes
- `git stash apply`: Aplica el stash más reciente
- `git stash drop`: Elimina el stash más reciente

**Ejemplos:**

````python
# Stash your current changes
!git stash

# List all stashes
!git stash list

# Apply the most recent stash
!git stash apply

# Drop the most recent stash
!git stash drop
````

## 13. Etiquetar Confirmaciones

Las etiquetas se usan para marcar puntos específicos en la historia como importantes (por ejemplo, lanzamientos).

- `git tag v1.0`: Crea una etiqueta simple
- `git tag -a v1.0 -m "Versión 1.0"`: Crea una etiqueta anotada
- `git tag`: Lista todas las etiquetas
- `git push origin v1.0`: Sube una etiqueta al repositorio remoto

**Ejemplos:**

````python
# Create a lightweight tag
!git tag v1.0

# Create an annotated tag
!git tag -a v1.0 -m "Version 1.0"

# List all tags
!git tag

# Push a tag to the remote repository
!git push origin v1.0
````

## 14. Ver Diferencias

Usa `git diff` para ver los cambios entre confirmaciones, ramas o entre tu directorio de trabajo y el último commit.

- `git diff`: Muestra los cambios no preparados
- `git diff --staged`: Muestra los cambios preparados
- `git diff rama1..rama2`: Muestra diferencias entre ramas

**Ejemplos:**

````python
# Show unstaged changes
!git diff

# Show staged changes
!git diff --staged

# Show differences between two branches
!git diff main..feature-branch
````

## 15. Ignorar Archivos con .gitignore

Un archivo `.gitignore` le indica a Git qué archivos o carpetas debe ignorar. Esto es útil para excluir archivos temporales, artefactos de compilación o información sensible.

**Ejemplo de .gitignore:**
```
# Ignorar archivos de caché de Python
__pycache__/
*.pyc

# Ignorar archivos de entorno
.env

# Ignorar carpetas de compilación
build/
dist/
```

Coloca el archivo `.gitignore` en la raíz de tu repositorio. Puedes editarlo en cualquier momento para agregar o quitar patrones.
